```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("Wine dataset.csv")
```

```python
df.head()
```

```python
df=df[["class","Alcohol","Malic acid"]]
```

```python
df.head()
```

```python
sns.kdeplot(df["Alcohol"])
```

```python
sns.kdeplot(df["Malic acid"])
```

```python
sns.scatterplot(x=df['Alcohol'],y=df['Malic acid'],hue=df['class'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.drop('class',axis=1),df['class'],test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(x_train)
x_train_scaled=scaler.transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train_scaled=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_test_scaled=pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
np.round(x_train.describe())
```

```python
np.round(x_train_scaled.describe())
```

```python
x_train_scaled
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.scatter(x_train['Alcohol'],x_train['Malic acid'])
ax1.set_title("Befor scalling")
ax2.scatter(x_train_scaled['Alcohol'],x_train_scaled['Malic acid'])
ax2.set_title("After scalling")
plt.show()
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("Befor scalling")
sns.kdeplot(x_train['Alcohol'],ax=ax1)
sns.kdeplot(x_train['Malic acid'],ax=ax1)



ax2.set_title("After scalling")
sns.kdeplot(x_train_scaled['Alcohol'],ax=ax2)
sns.kdeplot(x_train_scaled['Malic acid'],ax=ax2)

plt.show()
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("distribution Befor scalling")
sns.kdeplot(x_train['Alcohol'],ax=ax1)




ax2.set_title("distribution After scalling")
sns.kdeplot(x_train_scaled['Alcohol'],ax=ax2)

plt.show()
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2,figsize=(12,5))
ax1.set_title("distribution Befor scalling")
sns.kdeplot(x_train['Malic acid'],ax=ax1)




ax2.set_title("distribution After scalling")
sns.kdeplot(x_train_scaled['Malic acid'],ax=ax2)

plt.show()
```
